- **import**
  - import를 사용하여 해당 모듈 전체를 import
- **from import**
  - 해당 모듈에서 특정한 타입만 import
- **\* import**
  - 해당 모듈 내 정의된 모든 것을 import
- **as**
  - 모듈 impoty 시 별명 지정 가능

```python
import math
math.pi
```

```python
math.cos(100)
```
